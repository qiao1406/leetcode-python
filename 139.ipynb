```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """

        path = []
        w = set()
        wordDict.sort(key=len)
        for word in wordDict:
            flag = True
            for i in w:
                if len(word) % len(i) == 0 and word == i * (len(word) // len(i)):
                    flag = False
                    break
            if flag:
                w.add(word)

        start, end = 0, 1

        while True:
            if s[start: end] in w:
                if end == len(s):
                    return True
                else:
                    path.append((start, end))
                    start = end
            elif end == len(s):
                if path:
                    start, end = path.pop()
                else:
                    return False
            end += 1
        
s = Solution()
# s.wordBreak('catsand', ['cats', 'cat', 'and'])
# s.wordBreak(s = "catsandog", wordDict = ["cats", "dog", "sand", "and", "cat"])
# s.wordBreak(s='aaaaaaaaaaaaaaaaab', wordDict=['a', 'aa', 'aaa'])
s.wordBreak("aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaab",
["a","aa","aaa","aaaa","aaaaa","aaaaaa","aaaaaaa","aaaaaaaa","aaaaaaaaa","aaaaaaaaaa"])
```
