```python
class Fraction(object):

    def __init__(self, a, b):
        self.a = a
        self.b = b

    @staticmethod
    def gcd(x, y):
        if x == 0 or y ==0:
            return 1
        if x < y:
            y, x = x, y

        while y > 0:
            x, y = y, x % y
        return x

    def __add__(self, other):
        x = self.a * other.b + self.b * other.a
        y = self.b * other.b
        r = self.gcd(x, y)
        return Fraction(x / r, y / r)

    def __sub__(self, other):
        other = Fraction(0 - other.a, other.b)
        return self.__add__(other)

    def __mul__(self, other):
        x = self.a * other.a
        y = self.b * other.b
        r = self.gcd(x, y)
        return Fraction(x // r, y // r)

    def __truediv__(self, other):
        other = Fraction(other.b, other.a)
        return self.__mul__(other)

    def val(self):
        if self.b == 0:
            return 0
        return self.a / self.b
    
    def __repr__(self):
        return f'Fraction({self.a}, {self.b})'


class Solution(object):

    def op_res(self, op, x, y):
        if op == '+':
            return x + y
        elif op == '-':
            return x - y
        elif op == '*':
            return x * y
        elif op == '/':
            if y.val() == 0:
                return Fraction(0, 0)
            return x / y

    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """

        ans = []
        n = len(nums)

        def backtrace(first=0):
            if first == n:
                ans.append(nums[:])
            else:
                for i in range(first, n):
                    nums[first], nums[i] = nums[i], nums[first]
                    backtrace(first + 1)
                    nums[first], nums[i] = nums[i], nums[first]
        
        backtrace()
        return ans

    def judgePoint24(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """

        nums = [Fraction(i, 1) for i in nums]
        ops = ['+', '-', '*', '/']
        permute_nums = self.permute(nums)
        
        for p in permute_nums:
            n1, n2, n3, n4 = p

            # case 1: ((n1 op1 n2) op2 n3) op3 n4
            for op1 in ops:
                val1 = self.op_res(op1, n1, n2)
                for op2 in ops:
                    val2 = self.op_res(op2, val1, n3)
                    for op3 in ops:
                        val3 = self.op_res(op3, val2, n4)
                        if val3.val() == 24:
                            return True
                            
            # case2 (n1 op1 n2) op2 (n3 op3 n4)
            for op1 in ops:
                left_val = self.op_res(op1, n1, n2)
                for op3 in ops:
                    right_val = self.op_res(op3, n3, n4)
                    for op2 in ops:
                        val = self.op_res(op2, left_val, right_val)
                        if val.val() == 24:
                            return True
            
            # case3 (n1 op1 (n2 op2 n3)) op3 n4
            for op2 in ops:
                val1 = self.op_res(op2, n2, n3)
                for op1 in ['/', '-']:
                    val2 = self.op_res(op1, n1, val1)
                    for op3 in ops:
                        val3 = self.op_res(op3, val2, n4)
                        if val3.val() == 24:
                            return True
            
            # case4 n1 op1 (n2 op2 (n3 op3 n4))
            for op3 in ops:
                val1 = self.op_res(op3, n3, n4)
                for op2 in ops:
                    val2 = self.op_res(op2, n2, val1)
                    for op1 in ['/', '-']:
                        val3 = self.op_res(op1, n1, val2)
                        if val3.val() == 24:
                            return True
            
        return False


s = Solution()
s.judgePoint24([1, 2, 1, 9])
# x = Fraction(7, 1)
# y = Fraction(1, 1)
# print(x - y)
```
